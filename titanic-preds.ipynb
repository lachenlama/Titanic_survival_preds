```python
df_train = pd.read_csv("../input/titanic/train.csv")
df_train.head()
```

```python
df_train.shape
```

```python
df_test = pd.read_csv("../input/titanic/test.csv")
df_test.head()
```

```python
df_test.shape
```

```python
women = df_train.loc[df_train.Sex == 'female']['Survived']
rate_women = sum(women)/len(women)
print("% of women's survival is ",rate_women)
```

```python
men = df_train.loc[df_train.Sex == 'male']['Survived']
rate_men = sum(men)/len(men)
print("% of men's survival is ", rate_men)
```

* As we can see, 75% of women survived from the Titanic crash

```python
plt.figure(figsize=(10,6))
sns.barplot(x=df_train.Sex, y=df_train.Survived)
```

Barplot of Survivals according to gender.

```python
from sklearn.ensemble import RandomForestClassifier
y = df_train['Survived']

features = ['Pclass','Sex', 'SibSp', 'Parch']
X = pd.get_dummies(df_train[features])
X_test = pd.get_dummies(df_test[features])

model = RandomForestClassifier(n_estimators = 100, max_depth=5, random_state=1)
model.fit(X,y)
prediction = model.predict(X_test)
```

```python
output = pd.DataFrame({'PassengerId': df_test.PassengerId, 'Survived':prediction})
output.to_csv('submission.csv', index = False)
print("Done")
```
